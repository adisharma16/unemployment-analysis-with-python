```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_corr = cleaned_df.corr(numeric_only = True)
```

```python
## plt fig
plt.figure(figsize =(6,6))
plt.title("correlation matix")
sns.heatmap(df_corr , annot = True)
plt.show()
```

insight :
Estimated Unemployment Rate and Estimated Employed show a weak negative correlation (-0.22), indicating that higher employment levels are generally associated with slightly lower unemployment rates.

Estimated Labour Participation Rate has almost no correlation with either unemployment rate (0.0026) or employment (0.011), suggesting a very weak linear relationship.

Overall, the correlation matrix indicates that the variables are largely independent, with no strong positive or negative relationships observed in the dataset

```python
plt.title("Unemployment by Region")
sns.barplot(x = "Region", y = "Estimated Umeployment Rate" , data = cleaned_df, palette = "winter")
plt.show()
```

```python
sns.scatterplot(
    x = "Estimated Employed",
    y = "Estimated Umeployment Rate",
    hue = "Region",
    data = cleaned_df
)
print(cleaned_df.columns)
```

#### Insights from the Data

Unemployment Rate Trend: Fell to 3.05% in June, then rose again, peaking at 5.17% in September, showing short‑term volatility in the rural labour market.

Employment Numbers: Gradually increased from ~11.99M in May to ~12.28M in August–September, reflecting steady rural job absorption.

Labour Participation Rate: Improved from 43.24% in May to 44.68% in September, but rising unemployment suggests job creation is slower than workforce entry.

```python
state_data = cleaned_df.groupby("Region")["Estimated Umeployment Rate"].mean().sort_values()
```

```python
plt.title("Distibution of Umeployment Rate")
plt.pie(
    state_data,
    labels = state_data.index,
    autopct = "%1.1f%%",
    startangle = 90,
)
plt.show()
```

#### Insights from the Chart
Urban vs Rural Gap: Urban areas account for 56% of unemployment, which is significantly higher than rural areas at 44%.

Urban Concentration: The higher share in urban regions suggests that unemployment challenges are more concentrated in cities, possibly due to migration and limited job absorption.

Rural Advantage: Rural areas show relatively lower unemployment, indicating stronger labour participation or more informal job opportunities compared to urban regions.

### Final Conclusion from the Project
Rural vs Urban Divide: Urban regions consistently show higher unemployment compared to rural areas, highlighting structural challenges in city job markets.

Volatility in Rural Employment: Rural unemployment rates fluctuate seasonally, but overall employment numbers and labour participation show gradual improvement.

COVID‑19 Impact: Both rural and urban unemployment spiked sharply in mid‑2020, with urban areas hit harder, reflecting vulnerability of formal and urban jobs during crises.

Policy Implication: Strengthening rural job creation and stabilizing urban employment opportunities are crucial to balance workforce participation and reduce unemployment disparities.
